```python
import pandas as pd
posições=[]
coordenadas=[]


for alt in range(1,4):
    for larg in range(1,4):
        for prof in range(1,4):

            posições.append({'altura':alt,'largura':larg,'profundidade':prof})
            
cubo=pd.DataFrame(posições)


cubo.loc[cubo['altura']==3,'U']='amarelo'
cubo.loc[cubo['largura']==1,'L']='verde'
cubo.loc[cubo['profundidade']==1,'F']='laranja'
cubo.loc[cubo['largura']==3,'R']='azul'
cubo.loc[cubo['profundidade']==3,'B']='vermelho'
cubo.loc[cubo['altura']==1,'D']='branco'


for i in range(len(cubo)):
    coordenadas.append([cubo['altura'][i],cubo['largura'][i],cubo['profundidade'][i]])

for j in coordenadas:
    
    if j.count(2)==3:
        cubo.loc[coordenadas.index(j),'tipo']='nucleo'
        
    elif j.count(2)==0:
        cubo.loc[coordenadas.index(j),'tipo']='quina'

    elif j.count(2)==2:
        cubo.loc[coordenadas.index(j),'tipo']='centro'
    
    else:
        cubo.loc[coordenadas.index(j),'tipo']='meio'


cubo.drop(cubo.loc[cubo['tipo']=='centro'].index,axis=0,inplace=True)
cubo.drop(cubo.loc[cubo['tipo']=='nucleo'].index,axis=0,inplace=True)

cubo.reset_index(inplace=True,drop=True)


ciclo_U=cubo.loc[[14,16,19,18,17,15,12,13]]
ciclo_L=cubo.loc[[14,13,12,8,0,1,2,9]]
ciclo_F=cubo.loc[[12,15,17,10,5,3,0,8]]
ciclo_R=cubo.loc[[17,18,19,11,7,6,5,10]]
ciclo_B=cubo.loc[[19,14,16,9,2,4,7,11]]
ciclo_D=cubo.loc[[0,3,5,6,7,4,2,1]]



rotação=pd.concat([ciclo_L[2:],ciclo_L[:2]])


ciclo_L.reset_index(inplace=True)
rotação.reset_index(inplace=True)

display(ciclo_L)
display(rotação)

ciclo_L.loc[:,1:3]=rotação.loc[:,1:3]
```
