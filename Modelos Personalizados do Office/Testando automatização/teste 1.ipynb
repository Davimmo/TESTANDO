```python

listaA={'repolho':'vegetal','banco':'santander','endereço':'petrolina','nome':'geovana','musica':'ela partiu'}
listaB={'filme':'senhor do aneis','veiculo':'moto','linguagem':'python','cor':'amarelo','novela':'da cor do pecado'}

def troca_valores(listaA,listaB):
    novalista=list(listA.keys())

    valorespassados=list(listaB.values())

    for i in novalista[2:]:
        
        listaA[i]=(valorespassados[int(novalista.index(i))])

        return listaA

troca_valores(listaA,listaB)



    
```
