```python
import pandas as pd
from tika import parser
def achar_matricula(nome,pdf):
    posição=pdf.find(nome.upper())
    if posição!=-1:
        matricula=pdf[posição-13:posição-3]
        print(f'NOME:{nome};MATRICULA:{matricula}')
        return matricula
        

dicio=[]
funcionarios_df=pd.read_excel('Funcionarios.xlsx')

raw=parser.from_file('ESCOLA.pdf')
conteudo=raw['content']


for nome in funcionarios_df['NOME']:
    matricula=achar_matricula(nome,conteudo)
    dicio.append({'nome':nome,'matricula':matricula})


dicio=pd.DataFrame(dicio)












```
